# 第１章　データの整理

データは集めただけでは数が多すぎて、そこから何か読み取りことはできない。
大量のデータから何かを読み取るには文章のように短い言葉でまとめる、
つまり、大量のデータから短い数値（言葉）に要約してデータの性質を示してあげる必要がある。
その要約についての学問が記述統計学である。

では早速要約方法の説明に取り掛かりたいところだが、要約の前にあらかじめデータの種類について知っておく必要がある。
なぜデータの種類について知っておく必要があるかというと、データの種類によって適用できる要約方法が限定されるからだ。
たとえば、一般的に有名な*平均値*はデータ要約のひとつであり、身長や体重などデータには適用できる。しかし、
部署名や都道府県名は平均値を計算することができない。

## データの種類

データは物差しを使って記録される。その物差しのことを尺度と呼ぶ。尺度には身長や距離のような長さや、性別のような二値情報など様々な種類が存在する。
この尺度の種類に応じてデータが分類される。これを*測定尺度*または*尺度水準*と呼んでいる。

*尺度*は性質に応じて４種類に分けられる。それぞれ*名義尺度*と*順位尺度*、*間隔尺度*と*比率尺度*である。それぞれの尺度の性質は以下である。

* 名義尺度
  * 値の等価性が判断できる    
* 順位尺度
  * 名義尺度の性質を含む
  * 値の大小関係が判断できる
* 間隔尺度
  * 順位尺度の性質を含む
  * 値の差に意味がある
    * AとBの差とCとDの差が同じであれば、A-B間の間隔とC-D間の間隔は同じであると言うことができる
* 比率尺度
  * 間隔尺度の性質を含む
  * マイナスは存在しない
  * 絶対的なゼロ点を持つ（間隔尺度はゼロ点を自由に決めることができる）
    * 絶対的なゼロ点とは、その尺度でそれ以外のゼロ点が存在しないという意味である
    * 例えば、身長は0cmというゼロ点以外のゼロ点は存在しない
    * 絶対的なゼロ点を持つことにより、AはBの2倍であるという比較が可能になる

*名義尺度*と*順位尺度*、*比率尺度*と*間隔尺度*を使って測定されたデータはそれぞれ、*カテゴリデータ*と*順位データ*、*間隔データ*と*比率データ*に分けられる。それぞれデータは以下のような性質を持つ

* カテゴリデータ
  * 尺度: 名義尺度
  * 性質:
    * 適用できる演算:
      * 等価性判断
      * 度数カウント
    * 代表値: 最頻値
  * 例: 電話番号、性別、郵便番号    
* 順位データ
  * 尺度: 順位尺度
  * 性質:
    * 適用できる演算:
      * 等価性判断
      * 大小比較
    * 代表値: 中央値、最頻値
  * 例: 評価値、硬度
* 間隔データ
  * 尺度: 間隔尺度
  * 性質:
    * 適用できる演算:
      * 等価性判断
      * 大小比較
      * 加算、減算
    * 代表値: 算術平均、中央値、最頻値
  * 例: 温度、知能指数
* 比率データ
  * 尺度: 比率度
  * 性質:
    * 適用できる演算:
      * 等価性判断
      * 大小比較
      * 加算、減算、乗算、除算
    * 代表値: 各種平均値
  * 例: 身長、時間、お金
  
*カテゴリデータ*と*順位データ*は、まとめて*質的データ*と呼ばれ、*間隔データ*と*比率データ*はまとめて*量的データ*と呼ばれている。

## データを取得してからまずやること

データを取得し、そのデータの種類を分類したらすぐに要約に入るのではなく、データをグラフ化し、どんな感じにばらついているかを視覚的に判断できるようにします。これはデータを視覚的にとらえることによりデータの特徴を把握するためです。
もちろん、平均値は分散を使ってデータの特徴を捉えることは可能ですが、視覚的に分かる方がとっかかりやすいです。

視覚的にデータを把握するための手段として*度数分布表*や*ヒストグラム*があります。

*度数分布表*を作成する場合は以下の情報も作成するのが理想的です。

  * 階級
  
    データを分割した区間の値のこと: A～B, C～D
    
  * 階級値
  
    その階級（区間）の代表値。平均値や中央値、最頻値などが使われる
 
  * 度数
  
    度数はその階級の入るデータの数
    
  * 相対度数(可能であれば)
  
    度数が全体の何%かを表す値
    
  * 累積相対度数(可能であれば)
  
    相対度数を足し合わせたもの。最終的には100%になる
    
カテゴリデータの場合は、それぞれのカテゴリを区間として利用すれば良い。
順位データの場合も同様に、それぞれの値を区間として利用すれば良い（？）。 # <- 要確認
間隔データや比率データのような数値データの場合、適当な区間でデータを区切る必要がある。
区切る間隔に特に決まりはないが、「データの数の平方根＋１」だと割と良いらしい。


ここで、数値データについて、例えば身長や体重のように連続した値をとり得るものは *連続変量* と呼ばれている。
逆に人数や戸数のように離散的な値を取り得るものは *離散変量* と呼ばれている。

*変量* とは、試行前は不明だが施行後に値が観測されるような特性のことを *変量* と呼ぶ。
身長や体重のように連続した値をとり得る変量のことを *連続変量* と呼び、さいころの目の数や人数などのように
離散的な値をとるような変量のことを *離散変量* と呼ぶ。

ヒストグラムはカテゴリデータをグラフ化したものである。

## 代表値: 平均値

### 算術平均

データの特徴を数値で一気に表現したものを *統計量* と呼び、その中の一つに *代表値* というものがある。

*代表値*はいくつか存在ます。その中でもっとも有名なものは *平均値* である。

平均値はデータをすべて加算し、データの個数で割ることで求められる。

```
  平均値 = sum(x1, x2, ..., xn)/n
```
ここで、平均値と言っているものは *算術平均* と呼ばれている。

### 加重平均

度数分布表のように、階級値と度数しか表示されていなかったり、ある特定の分野の値の重みづけをして平均値を求めたい場合は
算術平均ではなく *加重平均* というものを使う。

*加重平均* は、各値に重みを掛け、それらの総和に対し重みづけの総和で割った値である。
```
加重平均 = (w1*x1 + w2*x2 + ... wn*xn)/(w1 + w2 + ... + wn)
```

### 幾何平均

掛け合わせた値の平均伸び率のようのな値を求めるためには *幾何平均* を利用する。
*幾何平均* は、掛け算した値に対しかけた数の累乗根を求めることで算出できる。

*累乗根*は成長率のようなものの計算に用いられることが多い。

* 参照: http://oto-suu.seesaa.net/article/183025125.html

*幾何*とついているのは図形が関係しているからではなく、数学では掛け算する計算に対して*幾何*とつけることが多いことに由来しているらしい。

### 移動平均

その他の平均値として*移動平均*がある。
*移動平均*は、株価や気象データなど時系列データを平滑化するための手法で、個々のイレギュラーな値に影響されずにトレンドなどを
つかむことができるようになす。

移動平均は何種類か存在するが、一番簡単なのが単純移動平均である。
単純移動平均は、適当に決めた日数分のデータの加算平均値をそれぞれの時点で求めていけば良い。
短期的なトレンドを知りたい場合は、日数を短くい、長期的なトレンドをつかみたい場合は長く設定する。

ただし、長期的なトレンドを単純移動平均でつかもうと思ってもあまり使えない場合が多いので、他の移動平均があるが
ここでは割愛する。


## 代表値: ばらつき

平均値だけの場合、データのばらつきの情報が消えてしまう。
データのばらつきを表すための情報として*偏差*がある。

### 偏差

偏差は、それぞれの値を算術平均値で引いた値のことである。
これで各値が算術平均値からどれくらい離れているかが分かる。

しかし、例えば二つのグループのどちらがデータの散らばり具合が大きいかが分からない。

そこで、偏差の平方根を足した値（*偏差平方和*）が用いられる。平方にする理由は、値にマイナスが含まれている場合があり、
偏差同士で打ち消し合ってしまい情報が消えてしまうからである。

*偏差平方和*を用いれば同じ種類のグループ間のデータの散らばりぐらいを比較することができる。
値が大きい方がより散らばっている。

### 分散

しかし、データの数が多くなると*偏差平方和*の値がとてつもなく大きくなってしまうため、
*偏差平方和*をデータ数で割った値（*分散*）が用いられる。

### 標準偏差

さらに、*分散*は元のデータの平方根になっており値の単位も元データの平方になっている。そのため、
*分散*の平方根の値（*標準偏差*）がばらつきの指標としてよく使われる。

### 変動係数

*標準偏差*は同じ種類のグループ間のばらつきの比較に使う分には問題がないが、種類の違うグループの間（例えばクジラの体重と人の体重のデータなど）のばらつきを比較するには用いることができない（そもそもクジラの体重は人間の体重と比べ物にならない程大きいためちょっとした差が人間にとって大きな差になる）。

この問題を解決するために、*標準偏差*をグループの*算術平均*で割った値（*変動係数*）が使われる。
*変動係数*は単位が消失すた値（*無名数*）ため、種類の異なるグループのばらつきの比較が可能になる。

## 質的データの場合の代表値について

質的データである順位データやカテゴリデータの場合は上記の平均値や分散を求めることができないが、要約できないわではない。

順位データの場合、平均値の代わりに*中央値(メジアン)*や*最頻値(モード)*が用いられる。最頻値は度数分布表で一番度数が多い階級の階級値のことである。
カテゴリデータの場合は、順序がないため中央値は使えず、*最頻値*しか使えない。

またばらつきは計算で求めることができないため、図やグラフにして視覚的に確認するしかない。

ただし、例外があり、順序データの場合は値に対し数値を割り当て(*コーディング*と呼ぶ）平均値を求めることがある。
ここで求めた平均値はそれぞれのグループの傾向をつかむために使われたりする。

話はそれるが、データを集めるときに適切にサンプリングを行わないとデータの方より(*サンプリングバイアス*）が起きてしまうので注意が必要。

# 相関係数

上記の平均値や分散は１つの変量についての要約値の説明であったが、二つの変量に対して、どういう関係があるかを説明する統計量がある。
*ピアソンの積率相関係数*という値は二つの変量が線形的に相関があるかどうかを説明するための変数である。
この値は-1～1の間の値をとり、1に近ければ正の相関があるといい、-1に近ければ負の相関があるという。0に近い場合は相関がないという。

ExcelでCORREL関数を使うことにより相関値を求めることができる。
