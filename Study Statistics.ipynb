# 第１章　データの整理

データは集めただけでは数が多すぎて、そこから何か読み取りことはできない。
大量のデータから何かを読み取るには文章のように短い言葉でまとめる、
つまり、大量のデータから短い数値（言葉）に要約してデータの性質を示してあげる必要がある。
その要約についての学問が記述統計学である。

では早速要約方法の説明に取り掛かりたいところだが、要約の前にあらかじめデータの種類について知っておく必要がある。
なぜデータの種類について知っておく必要があるかというと、データの種類によって適用できる要約方法が限定されるからだ。
たとえば、一般的に有名な*平均値*はデータ要約のひとつであり、身長や体重などデータには適用できる。しかし、
部署名や都道府県名は平均値を計算することができない。

## データの種類

データは物差しを使って記録される。その物差しのことを尺度と呼ぶ。尺度には身長や距離のような長さや、性別のような二値情報など様々な種類が存在する。
この尺度の種類に応じてデータが分類される。これを*測定尺度*または*尺度水準*と呼んでいる。

*尺度*は性質に応じて４種類に分けられる。それぞれ*名義尺度*と*順位尺度*、*間隔尺度*と*比率尺度*である。それぞれの尺度の性質は以下である。

* 名義尺度
  * 値の等価性が判断できる    
* 順位尺度
  * 名義尺度の性質を含む
  * 値の大小関係が判断できる
* 間隔尺度
  * 順位尺度の性質を含む
  * 値の差に意味がある
    * AとBの差とCとDの差が同じであれば、A-B間の間隔とC-D間の間隔は同じであると言うことができる
* 比率尺度
  * 間隔尺度の性質を含む
  * マイナスは存在しない
  * 絶対的なゼロ点を持つ（比較尺度はゼロ点を自由に決めることができる）
    * 絶対的なゼロ点とは、その尺度でそれ以外のゼロ点が存在しないという意味である
    * 例えば、身長は0cmというゼロ点以外のゼロ点は存在しない
    * 絶対的なゼロ点を持つことにより、AはBの2倍であるという比較が可能になる

*名義尺度*と*順位尺度*、*比率尺度*と*間隔尺度*を使って測定されたデータはそれぞれ、*カテゴリデータ*と*順位データ*、*間隔データ*と*比率データ*に分けられる。それぞれデータは以下のような性質を持つ

* カテゴリデータ
  * 尺度: 名義尺度
  * 性質:
  　* 適用できる演算:
      * 等価性判断
      * 度数カウント
    * 代表値: 最頻値
  * 例: 電話番号、性別、郵便番号    
* 順位データ
  * 尺度: 順位尺度
  * 性質:
    * 適用できる演算:
      * 等価性判断
      * 大小比較
    * 代表値: 中央値、最頻値
  * 例: 評価値、硬度
* 間隔データ
  * 尺度: 間隔尺度
  * 性質:
    * 適用できる演算:
      * 等価性判断
      * 大小比較
      * 加算、減算
    * 代表値: 算術平均、中央値、最頻値
  * 例: 温度、知能指数
* 比率データ
  * 尺度: 比率度
  * 性質:
    * 適用できる演算:
      * 等価性判断
      * 大小比較
      * 加算、減算、乗算、除算
    * 代表値: 各種平均値
  * 例: 身長、時間、お金
  

まずデータにはいくつか種類がある。それは数値的に扱えるものと扱えないものである。それぞれ*量的データ*、*質的データ*という。
例えば、身長や体重、年収などは*量的データ*であり、都道府県や部署名などは*質的データ*である。


さらに、*量的データ*と*質的データ*はそれぞれ二種類に分類できる。それが、*比率データ*と*間隔データ*、*順位データ*と*カテゴリデータ*である。
それぞれ以下のような性質を持っている。

| データ分類名   | 意味のある演算 | 例                       |
|:--------------:|:--------------:|:------------------------:|
| 比率データ     | + - * / > =    |身長、体重、時間、金額    |
| 間隔データ     | + - > =        | 温度（摂氏）、知能指数   |
| 順序データ     | > =            | 満足度、硬度、評価値     |
| カテゴリデータ | =              | 電話番号、郵便番号、性別 |
