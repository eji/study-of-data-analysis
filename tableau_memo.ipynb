# Tableau Softwareの使い方についてのメモ

## あるフィールドのユニークな値の数の合計

1. 分析(A) > 計算フィールドの作成(C)を選ぶ

    計算フィールドウィンドウが表示される

2. 名前に適当な名前を付ける（後にメジャーに表示されるようになる）

3. 関数(U)から**COUNTD**を選択し、引数にフィールドの名前を選択し、OKボタンを押下

    メジャーに作成した計算フィールドが追加される

4. メジャーから作成した計算フィールドをDrag&Dropでシートに移すことでカウント値が表示されるようになる
